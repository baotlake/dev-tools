```python
code_list = ['ar', 'am', 'bg', 'bn', 'ca', 'cs', 'da', 'de', 'el', 'en', 'en_GB', 'en_US', 'es', 'es_419', 'et', 'fa', 'fi', 'fil', 'fr', 'gu', 'he', 'hi', 'hr', 'hu', 'id', 'it',
             'ja', 'kn', 'ko', 'lt', 'lv', 'ml', 'mr', 'ms', 'nl', 'no', 'pl', 'pt_BR', 'pt_PT', 'ro', 'ru', 'sk', 'sl', 'sr', 'sv', 'sw', 'ta', 'te', 'th', 'tr', 'uk', 'vi', 'zh_CN', 'zh_TW']
```

```python
from os import path
import json
```

```python
json.loads('{}')
```

```python

output_dict = json.load(open('output.json'))
# print(output_dict)

for c in code_list:
    file_path = path.join('./_locales/', c, 'messages.json')
    with open(file_path, 'r+', encoding='utf-8-sig') as f:
        text = f.read()
        message = json.loads(text)
        message["extName"]["message"] = output_dict[c]["name"]
        message["extDesc"]["message"] = output_dict[c]["des"]
        # print(message)
        # print(json.dumps(message, indent=4, ensure_ascii=False))
        f.seek(0)
        f.write(json.dumps(message, indent=4, ensure_ascii=False))
        f.truncate()

```
